Importo el datasset de Agricultura, el cual tienen indicadores de rendimiento de cultivos y demas

```python
df_muj = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/proyecto/desempleo_mujeres.csv", skiprows=4, on_bad_lines='skip',index_col=-1)

df_muj
```

Se programa la funcion "get_years" para obtener una lista de los years en tipo string

```python
def get_years():
  years = []
  for i in range(1960,2022):
    years.append(str(i))
  return years
```

Se usa la funcion melt para convertir los years de filas a una sola columna

```python

dfak = df_muj.melt(id_vars=['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code']
               ,value_vars=get_years(),var_name='Year',value_name='Valor')
dfak['Year'] = dfak['Year'].astype(int)
```

Quitamos la columna Indicator Name del dataframe, dejando solo los codigos de indicador

```python
dfak = dfak[['Country Name', 'Country Code', 'Indicator Code','Year', 'Valor']]
```

Usamos la funcion 'fillna' para quitar los valores 'NaN' del dataframe

```python
dfak.fillna(0.,inplace=True)
```

Hacemos 'pivot' para pasar los codigos de indicadores de Columna a filas

```python
dfak = dfak.pivot(index=['Country Name', 'Country Code','Year'],columns=[ 'Indicator Code']
                    ,values='Valor').reset_index()
```

filtro del dataframe solo los indicadores que me interesan

```python
dfak = dfak[['Country Name', 'Country Code','Year','SL.UEM.TOTL.FE.ZS']]
```

```python
dfak.sample(5)
```
